```python
df['stipend'] = df['stipend'].apply(lambda x: '0' if 'Unpaid' in str(x) else x)
df['stipend'] = df['stipend'].str.replace("/month", "").str.replace("lump sum", "").str.replace("₹", "")
df['Other Benifits'] = df['stipend'].str.split('+').apply(lambda x: x[-1].strip() if len(x) > 1 else None)
df['stipend'] = df['stipend'].str.split('+').str[0].str.strip()
```

```python
plt.figure(figsize=(15, 6))
sns.countplot(y='stipend', data=df, order=df['stipend'].value_counts().index[:10],palette="Set2")
plt.title('Top 10 Stipend Ranges')
plt.xlabel('Number of Counts')
plt.ylabel('Internship Stipend')
plt.show()
```

```python
import plotly.express as px
# Clean and extract numerical values from the stipend column
df['stipend'] = df['stipend'].replace('[\₹,/month]', '', regex=True)
df['stipend'] = df['stipend'].astype(str)
df['stipend'] = df['stipend'].apply(lambda x: x.split('-')[0].strip() if '-' in x else x.strip())
df['stipend'] = pd.to_numeric(df['stipend'], errors='coerce')

fig = px.histogram(df, x='stipend', nbins=30, title='Stipend Distribution', labels={'stipend': 'Stipend (INR/month)', 'count': 'Frequency'})
fig.update_traces(marker_color='skyblue', opacity=0.7)
fig.update_layout(bargap=0.1, bargroupgap=0.4, xaxis_title='Stipend (INR/month)', yaxis_title='Frequency')
fig.show()
```

```python
# Calculate average stipend for each company
average_stipend = df.groupby('company_name')['stipend'].mean().reset_index(name='average_stipend')

# Identify top-paying companies
top_paying_companies = average_stipend.sort_values(by='average_stipend', ascending=False).head(10)

# Create a bar chart
plt.figure(figsize=(12, 5))
sns.barplot(x='company_name', y='average_stipend', data=top_paying_companies, palette='viridis',width=0.5)
plt.title('Highest Internship Stipend Paying Companies\n')
plt.ylabel('Internship Stipend')
plt.xlabel('Companies Name')
plt.xticks(rotation=90)
# plt.grid(True)
plt.show()
```
